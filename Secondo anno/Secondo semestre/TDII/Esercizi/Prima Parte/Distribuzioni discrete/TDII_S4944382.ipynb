```python
## Distribuzione discreta Binomiale ## 
## Cattaneo Kevin - S4944382 - TDII - Informatica UniGE ##

from scipy.stats import binom # libreria per calcolare binomiale
import matplotlib.pyplot as plt # libreria per i grafici
import numpy as np

# Variabili n e p
n = 30
p = 0.7

# Indice della sommatoria i
x = list(range(n + 1))

# Calcolo i valori di pmf e cdf in ogni punto
pmf = binom.pmf(x, n, p)
cdf = np.cumsum(pmf)

# Divido il plot
figure, axis = plt.subplots(2)

# Plot del grafico PMF
axis[0].scatter(x, pmf, label="PMF", color="blue", marker="o", s=50)
axis[0].vlines(x, 0, pmf, colors='b', lw=2, alpha=0.4)
axis[0].set_xlim(0,n)
axis[0].set_ylim(0,0.2)
axis[0].set_xlabel("i-th value")
axis[0].set_ylabel("Probability Values")
axis[0].set_title("PMF for discrete distribution")

# Plot del grafico CDF
axis[1].plot(x,cdf,marker="o",label="CDF", color= "red")
axis[1].vlines(x, 0, cdf, colors='r', lw=2, alpha=0.4)
axis[1].set_xlim(0,n)
axis[1].set_ylim(0,1)
axis[1].set_xlabel("i-th value")
axis[1].set_ylabel("Probability Values")
axis[1].set_title("CDF for discrete distribution")

plt.tight_layout()
plt.show()
```

```python
## Distribuzione discreta Geometrica ## 
## Cattaneo Kevin - S4944382 - TDII - Informatica UniGE ##

from scipy.stats import geom # libreria per calcolare geometrica
import matplotlib.pyplot as plt # libreria per i grafici
import numpy as np

# Variabili n e p
n = 35
p = 0.1

# Indice della sommatoria i
x = list(range(n + 1))

# Calcolo i valori di pmf e cdf in ogni punto
pmf = geom.pmf(x,p)
cdf = np.cumsum(pmf)

# Divido il plot
figure, axis = plt.subplots(2)

# Plot del grafico PMF
axis[0].scatter(x, pmf, label= "PMF", color= "blue", marker= "o", s=50)
axis[0].vlines(x, 0, pmf, colors='b', lw=2, alpha=0.4)
axis[0].set_xlim(0,n)
axis[0].set_ylim(0,0.1)
axis[0].set_xlabel("i-th value")
axis[0].set_ylabel("Probability Values")
axis[0].set_title("PMF for discrete distribution")

# Plot del grafico CDF
axis[1].plot(x,cdf,marker="o",label="CDF", color= "red")
axis[1].vlines(x, 0, cdf, colors='r', lw=2, alpha=0.4)
axis[1].set_xlim(0,n)
axis[1].set_ylim(0,1)
axis[1].set_xlabel("i-th value")
axis[1].set_ylabel("Probability Values")
axis[1].set_title("CDF for discrete distribution")

plt.tight_layout()
plt.show()
```

```python
## Distribuzione discreta Poissoniana ## 
## Cattaneo Kevin - S4944382 - TDII - Informatica UniGE ##

from scipy.stats import poisson # libreria per calcolare poisson
import matplotlib.pyplot as plt # libreria per i grafici
import numpy as np

# Variabili n e p
n = 20  # nota: n rappresenta i seguendo la consegna fornita
p = 0.1
mu = 12

# Indice della sommatoria i, in accordo a quanto detto sopra
x = list(range(n + 1))

# Calcolo i valori di pmf e cdf in ogni punto
pmf = poisson.pmf(x,mu)
cdf = np.cumsum(pmf)

# Divido il plot
figure, axis = plt.subplots(2)

# Plot del grafico PMF
axis[0].scatter(x, pmf, label= "PMF", color= "blue", marker= "o", s=50)
axis[0].vlines(x, 0, pmf, colors='b', lw=2, alpha=0.4)
axis[0].set_xlim(0,n)
axis[0].set_ylim(0,0.2)
axis[0].set_xlabel("i-th value")
axis[0].set_ylabel("Probability Values")
axis[0].set_title("PMF for discrete distribution")

# Plot del grafico CDF
axis[1].plot(x,cdf,marker="o",label="CDF", color= "red")
axis[1].vlines(x, 0, cdf, colors='r', lw=2, alpha=0.4)
axis[1].set_xlim(0,n)
axis[1].set_ylim(0,1)
axis[1].set_xlabel("i-th value")
axis[1].set_ylabel("Probability Values")
axis[1].set_title("CDF for discrete distribution")

plt.tight_layout()
plt.show()
```
